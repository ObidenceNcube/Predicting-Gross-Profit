```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
revenue_mean = 170
revenue_std = 20
iterations = 1000
```

```python
future_revenue = np.random.normal(revenue_mean, revenue_std, iterations)
future_revenue
```

```python
plt.figure(figsize = (15,8))
plt.plot(future_revenue)
plt.show()
```

```python
COGS = -(future_revenue * np.random.normal(0.6,0.1))

plt.figure(figsize = (15,8))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
Gross_Profit = future_revenue + COGS
Gross_Profit
plt.figure(figsize = (15,8))
plt.plot(Gross_Profit)
plt.show
```
